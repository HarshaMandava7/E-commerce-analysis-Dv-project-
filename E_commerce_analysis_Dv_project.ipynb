```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.DataFrame({
    'Name': ['BigBasket', 'Shopcules', 'Myntra', 'Ajio', 'Flipkart', 'Amazon'],
    'Value': [13, 11, 7, 4, 2, 1]
})


circles = circlify.circlify(
    df['Value'].tolist(), 
    show_enclosure=False, 
    target_enclosure=circlify.Circle(x=0, y=0, r=1)
)

# Create just a figure and only one subplot
fig, ax = plt.subplots(figsize=(10,10))


# Remove axes
ax.axis('off')

# Find axis boundaries
lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

# list of labels
labels = df['Name']

# print circles
for circle, label in zip(circles, labels):
    x, y, r = circle
    ax.add_patch(plt.Circle((x, y), r, alpha=0.2, linewidth=2,color='blue'))
    plt.annotate(
          label, 
          (x,y ) ,
          va='center',
          ha='center',
          size=12
     )
```
